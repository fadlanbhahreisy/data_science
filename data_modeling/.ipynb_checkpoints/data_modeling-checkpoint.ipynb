**2.2 Join semua tabel**

```python
query = '''
        select *
        from Transaksi_snow Transaksi
            inner join Pelanggan_snow Pelanggan
                on (Transaksi.id_pelanggan = Pelanggan.id_pelanggan)
            inner join Sales_snow Sales
                on (Transaksi.id_sales = Sales.id_sales)
            inner join Kasir_snow Kasir
                on (Transaksi.id_kasir = Kasir.id_kasir)
            inner join (select * from Barang_snow Barang
            join Kategori_snow Kategori
            on (Barang.id_kategori = Kategori.id_kategori)) Barang
            on (Transaksi.id_barang  = Barang.id_barang)
        
'''
pysqldf(query)
```

# 3. Galaxy Schema

**3.1 Read dataset**

```python
Transaksi_galaxy = pd.read_excel("Transaksi_galaxy.xlsx",sheet_name='transaksi')
Pelanggan_galaxy = pd.read_excel("Transaksi_galaxy.xlsx",sheet_name='pelanggan')
Barang_galaxy = pd.read_excel("Transaksi_galaxy.xlsx",sheet_name='barang')
Sales_galaxy = pd.read_excel("Transaksi_galaxy.xlsx",sheet_name='sales')
Kasir_galaxy = pd.read_excel("Transaksi_galaxy.xlsx",sheet_name='kasir')
Kategori_galaxy = pd.read_excel("Transaksi_galaxy.xlsx",sheet_name='kategori')
Pesan_galaxy = pd.read_excel("Transaksi_galaxy.xlsx",sheet_name='pesan')
```

**3.2 Join Table barang dan Kategori**

```python
query = '''
        select Barang.nama_barang, Kategori.keterangan
        from Barang_galaxy Barang
        inner join Kategori_galaxy Kategori
        on (Barang.id_kategori = Kategori.id_kategori)
'''
pysqldf(query)
```

**3.3 Join table pelanggan, pesan dan sales**

```python
query = '''
        select Pesan.tanggal, Sales.nama_sales, Pelanggan.nama_pelanggan
        from Pesan_galaxy Pesan
        inner join Sales_galaxy Sales
        on (Pesan.id_sales = Sales.id_sales)
        inner join Pelanggan_galaxy Pelanggan
        on (Pesan.id_pelanggan = Pelanggan.id_pelanggan)
'''
pysqldf(query)
```

**3.4 Join table Transaksi dan kairr**

```python
query = '''
        select Transaksi.total, Transaksi.bayar, Transaksi.kembali, Kasir.nama_kasir
        from Transaksi_galaxy Transaksi
        inner join Kasir_galaxy Kasir
        on (Transaksi.id_kasir = Kasir.id_kasir)
'''
pysqldf(query)
```

**3.5 join all table**

```python
query = '''
        select Pesan.nama_pelanggan, Pesan.nama_sales, Transaksi.nama_kasir,
        Barang.nama_barang, Barang.keterangan, Pesan.tanggal,
        Transaksi.total, Transaksi.bayar, Transaksi.kembali
        from (select *
        from Transaksi_galaxy Transaksi
        inner join Kasir_galaxy Kasir
        on (Transaksi.id_kasir = Kasir.id_kasir)) Transaksi
        inner join (select *
        from Pesan_galaxy Pesan
        inner join Sales_galaxy Sales
        on (Pesan.id_sales = Sales.id_sales)
        inner join Pelanggan_galaxy Pelanggan
        on (Pesan.id_pelanggan = Pelanggan.id_pelanggan)) Pesan
        on (Transaksi.id_pesan = Pesan.id_pesan)
        inner join (select *
        from Transaksi_galaxy Transaksi
        inner join (select *
        from Barang_galaxy Barang
        inner join Kategori_galaxy Kategori
        on (Barang.id_kategori = Kategori.id_kategori)) Barang
        on (Transaksi.id_barang = Barang.id_barang)) Barang
        on (Transaksi.id_barang = Barang.id_barang)
        
'''
pysqldf(query)
```
